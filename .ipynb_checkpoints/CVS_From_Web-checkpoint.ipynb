```python
import datetime
import pandas as pd

#Creates API call to Urban Observatory, returns dataframe
def getFromUrbanObs(sensor, startTime, endTime):
    
    APIcall = getURL(sensor) + "startTime=" + startTime.isoformat() + "z&endTime=" + endTime.isoformat() + "z&outputAs=json"
    data = pd.read_json(APIcall)
    
    data = pd.DataFrame.from_records(data.historic[12])
    
    #data.set_index('time', inplace = True)
    #data.index = pd.to_datetime(data.index)
    
    return data
```

```python
import datetime
import pandas as pd
from datetime import timedelta

startTime = datetime.datetime(2019, 1, 1)
endTime = datetime.datetime(2021, 6, 1)

data = pd.DataFrame()

sensorName = "Lysimeter10_Net_radiation"

#Splits API calls into 7 days chucks to avoid time-out failure
tempEndTime = startTime
while tempEndTime < endTime:
    #set startTime & tempEndTime to next time-step
    startTime = tempEndTime
    tempEndTime = tempEndTime + timedelta(days = 3)
    
    #Create API call, append data to dataframe 'data'
    data = data.append(getFromUrbanObs(sensorName, startTime, tempEndTime))
    print("Collected up to " + tempEndTime.strftime("%d %b, %Y"))

print("Collected everything, finished!")

#formate data, order dataframe
data.set_index('time', inplace = True)
data.index = pd.to_datetime(data.index)
data = data.sort_values(by='time')

#Rename 'value' to sensor name
data[sensorName] = data['value']
data= data.drop(['value'], axis = 1)

data = data.drop(['duration'], axis = 1)

data.to_csv("CSV\\Lysimeter10_Weather\\" + sensorName + '.csv')
print(data)
```
